```python
import folium

folium.Map()   # world map
```

```python
folium.Map(location=[16.7850, 80.8488])
```

```python
from branca.element import Figure
fig = Figure(width = 350 ,height = 250)
map1 = folium.Map(location=[16.7850, 80.8488],zoom_start = 10,min_zoom = 5,max_zoom = 12)
fig.add_child(map1)
map1
```

```python
folium.TileLayer('Stamen Terrain').add_to(map1)
folium.TileLayer('Stamen Toner').add_to(map1)
folium.TileLayer('Stamen Water Color').add_to(map1)
folium.TileLayer('cartodbpositron').add_to(map1)
folium.TileLayer('cartodbdark_matter').add_to(map1)
folium.LayerControl().add_to(map1)
map1
```

```python
# adding markers
fig3=Figure(width=550,height=350)
map0=folium.Map(location=[16.7850, 80.8488],tiles='cartodbpositron',zoom_start=8)
fig3.add_child(map0)

folium.Marker(location=[16.5062, 80.6480],popup = 'Default popup Marker1',tooltip = 'click here to see more details',).add_to(map0)
folium.Marker(location=[16.7107, 81.0952],popup='<strong>Marker3</strong>',toolpit='<strong>Click here to see more popup</strong>').add_to(map0)
folium.Marker(location=[16.3067, 80.4365],popup='<h3 style="color:green;">marker 2</h3>',toolpit='<strong>Click here to see Guntur</strong>').add_to(map0)
```

```python
# Creating Basemap
fig3=Figure(width=550,height=350)
map3=folium.Map(location=[16.7850, 80.8488],tiles='cartodbpositron',zoom_start=8)
fig3.add_child(map3)

#Adding markers to the map
folium.Marker(location=[16.5062, 80.6480],popup='Default popup Marker1',tooltip='Click here to see more details',icon = folium.Icon(color='red',icon='none')).add_to(map3)
folium.Marker(location=[16.7107, 81.0952],popup='<strong>Marker3</strong>',tooltip='<strong>Click here to see Popup</strong>',icon = folium.Icon(color='green',prefix='glyphicon',icon='off')).add_to(map3)
folium.Marker(location=[16.3067, 80.4365],popup='<h3 style="color:green;">Marker2</h3>',tooltip='<strong>Click here to see Guntur</strong>',icon = folium.Icon(color='purple',prefix='fa',icon='anchor')).add_to(map3)

map3
```
