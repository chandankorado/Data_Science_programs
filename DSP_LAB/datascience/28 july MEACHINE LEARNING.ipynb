```python
# Machine Learning (mean, median, mode, standard deviation: std, Variance: var, Percentile: percentile )

import numpy as np                 # numpy for mean,median
from scipy import stats as sc      # scipy / stats for mode

speed = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

x = np.mean(speed)          # mean
print("\nMean of Speed : ",x)

y = np.median(speed)        # median
print("\nMedian of Speed : ",y)

z = sc.mode(speed)         # mode
print("\n",z)

s = np.std(speed)          # standard deviation
print("\nstandard deviation : ",s)

v = np.var(speed)          # Variance 
print("\nVariance : ",v)

p = np.percentile(speed, 90)   # Percentile
print("\nPercentile with 90 : ",p)

q = np.percentile(speed, 100)
print("\nPercentile with 100 : ",q)
```

```python
# Normal Data Distribution

import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(5.0, 1.0, 10000)   # mean is 5.0,  standard deviation is 1.0

plt.hist(x, 1000)
plt.show()
```

```python
# Random Data Distribution

import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(5.0, 1.0, 1000)
y = np.random.normal(10.0, 2.0, 1000)

plt.scatter(x, y,color = "hotpink")
plt.show()
```
