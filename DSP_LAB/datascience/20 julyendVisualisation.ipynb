```python
#  marker | line | color

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,5,3,7,3,1,6,0])
ypoints = np.array([3,1,6,0,1,5,3,7,])

plt.plot(xpoints,ypoints, 'o:r') # dooted line : red r
plt.show()
```

```python
plt.plot(xpoints,ypoints, 'o-g') # solid line - green g
plt.show()
```

```python
plt.plot(xpoints,ypoints, 'o--b') # Dashed line -- blue b
plt.show()
```

```python
plt.plot(xpoints,ypoints, 'o-.y') # dashed / dotted line -. yellow y
plt.show()
```

```python
plt.plot(xpoints,ypoints, 'o-k') # solid line - black k
plt.show()
```

```python
plt.plot(xpoints,ypoints, 'o-c') # solid line - cyan c
plt.show()
```

```python
plt.plot(xpoints,ypoints, 'o-m') # solid line - magenta
plt.show()
```

```python
plt.plot(xpoints,ypoints, 'o-g', ms = 20) # solid line - green g 
plt.show()                                # ms marker: size 
```

```python
plt.plot(xpoints,ypoints, 'o-y', ms = 20, mec = 'k') # solid line - black k
plt.show()                                           # mec : marker edge color 
```

```python
plt.plot(xpoints,ypoints, 'o-', ms = 20, mec = 'k', mfc = 'g') # solid line - black k
plt.show()
```

```python
plt.plot(xpoints,ypoints, 'o-', ms = 20, mec = 'k', mfc = 'hotpink') # solid line - black k
plt.show()
```

```python

plt.plot(xpoints,ypoints, 'o',linestyle = 'dashed', ms = 20, mec = 'k', mfc = 'hotpink') # solid line - black k
plt.show()                            # can also write like this
                                      # linestyle = 'dashed','dotted','solid','None','dashdot'
```

```python
# Compare Scatter plot

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,5,3,7,3,1,6,0])
ypoints = np.array([3,1,6,0,1,5,3,7,])

plt.scatter(xpoints,ypoints)

xpoints = np.array([1,5,6,0,3,7,3,1])
ypoints = np.array([3,3,7,1,6,0,1,5])

plt.scatter(xpoints,ypoints)

plt.show()
```

```python
# Compare Scatter plot with colors

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,5,3,7,3,1,6,0])
ypoints = np.array([3,1,6,0,1,5,3,7,])

plt.scatter(xpoints,ypoints, color = 'hotpink')

xpoints = np.array([1,5,6,0,3,7,3,1])
ypoints = np.array([3,3,7,1,6,0,1,5])

plt.scatter(xpoints,ypoints, color = 'k')

plt.show()
```

```python
# Scatter plot with different colors

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,5,3,7,3,1,6,0])
ypoints = np.array([3,1,6,0,1,5,3,7])

colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(xpoints,ypoints, c=colors)

plt.show()
```

```python
# Scatter plot with Color Map

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,5,3,7,3,1,6,0])
ypoints = np.array([3,1,6,0,1,5,3,7])

colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(xpoints,ypoints,c = colors,cmap = 'viridis')

plt.show()
```

```python
# Scatter plot with plt.colorbar()

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,5,3,7,3,1,6,0])
ypoints = np.array([3,1,6,0,1,5,3,7])

colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(xpoints,ypoints,c = colors,cmap = 'viridis')

plt.colorbar()

plt.show()
```

```python
# PIE

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,25,25,15])

plt.pie(y)

plt.show()
```

```python
# PIE with labels and start

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,25,25,15])
mylabels = ["Apple","Banana","Cherries","Dates"]

plt.pie(y, labels = mylabels,startangle = 0) # start from 0 degree angle

plt.show()
```

```python
# PIE with explode and shadow

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,25,25,15])
myexplode = [0.1, 0, 0, 0]
mylabels = ["Apple","Banana","Cherries","Dates"]



plt.pie(y, explode = myexplode,labels = mylabels,shadow = True)

plt.show()
```

```python
# PIE with colors

import matplotlib.pyplot as plt
import numpy as np

y = np.array([22, 20, 16,19,15,10,7,5])
myexplode = [0, 0, 0, 0.1,0,0,0,0]
mylabels = ["Red","Green","Blue","Cyan","Magenta","Yellow","Black","White"]
mycolors = ["red","green","blue","cyan","magenta","yellow","black","white"]



plt.pie(y, explode = myexplode,labels = mylabels,shadow = True,colors = mycolors)

plt.show()
```

```python
# pie with legends

import matplotlib.pyplot as plt
import numpy as np

y = np.array([22, 20, 16,19,15,10,7,5])
myexplode = [0, 0, 0, 0.1,0,0,0,0]
mylabels = ["Red","Green","Blue","Cyan","Magenta","Yellow","Black","White"]
mycolors = ["red","green","blue","cyan","magenta","yellow","black","white"]



plt.pie(y, explode = myexplode,labels = mylabels,shadow = True,colors = mycolors)
plt.legend()

plt.show()
```

```python
# pie with legends parameter

import matplotlib.pyplot as plt
import numpy as np

y = np.array([22, 20, 16,19,15,10,7,5])
myexplode = [0, 0, 0, 0.1,0,0,0,0]
mylabels = ["Red","Green","Blue","Cyan","Magenta","Yellow","Black","White"]
mycolors = ["red","green","blue","cyan","magenta","yellow","black","white"]



plt.pie(y, explode = myexplode,labels = mylabels,shadow = True,colors = mycolors)
plt.legend(title = "_Colors")

plt.show()
```
