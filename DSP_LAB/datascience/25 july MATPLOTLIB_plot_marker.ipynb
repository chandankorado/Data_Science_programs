```python
# Matplotlib

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,5,3,7])
ypoints = np.array([3,1,6,0])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
# plot with line

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,5,3,7])
ypoints = np.array([3,1,6,0])

plt.plot(xpoints,ypoints, marker = 'o')
plt.show()
```

```python
# plot without line

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,5,3,7])
ypoints = np.array([3,1,6,0])

plt.plot(xpoints,ypoints, 'o')
plt.show()
```

```python
# without x points 

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,1,6,0,2,1,7,5,8]) 

plt.plot(ypoints)               # x points will taken by default = 0, 1,2,3 4...
plt.show()
```

```python
# Marker

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,5,3,7])
ypoints = np.array([3,1,6,0])

plt.plot(xpoints,ypoints, '*')  # star *
plt.show()
```

```python
plt.plot(xpoints,ypoints, '.')   # point .
plt.show()
```

```python
plt.plot(xpoints,ypoints, ',')   # pixel .
plt.show()
```

```python
plt.plot(xpoints,ypoints, 'x')   # x
plt.show()
```

```python
plt.plot(xpoints,ypoints, 'X')   # X filled
plt.show()
```

```python
plt.plot(xpoints,ypoints, '+')   # + plus
plt.show()
```

```python
plt.plot(xpoints,ypoints, 'P')   # + filled
plt.show()
```

```python
plt.plot(xpoints,ypoints, 's')   # square  
plt.show()
```

```python
plt.plot(xpoints,ypoints, 'D')   # diamond 
plt.show()
```

```python
plt.plot(xpoints,ypoints, 'd')   # diamond thin
plt.show()
```

```python
plt.plot(xpoints,ypoints, 'p')   # pentagon 
plt.show()
```

```python
plt.plot(xpoints,ypoints, 'h')   # hexagon filled or H
plt.show()
```

```python
plt.plot(xpoints,ypoints, '^')   #  triangle up '^' triangle down 'v'
plt.show()
```

```python
 plt.plot(xpoints,ypoints, '>')   #  triangle right '>' triangle left '<' 
plt.show()
```

```python
 plt.plot(xpoints,ypoints, '1')   # tri up 1, down 2 , left 3, right 4
plt.show()
```

```python
 plt.plot(xpoints,ypoints, '_')   #  horizontal _ , vertical | 
plt.show()
```
